# Case Study Activity: Smart Door Access System for UMaT

You have been contacted by the **Dean of ORIC, UMaT**, to develop a smart door access system for UMaT admin staff. Each staff has an **access code** that must be protected from direct modification.

### Requirements:
Create a `Staff` class with:
*   **Public attribute:** `s_name`
*   **Private attribute:** `__access_code`
*   **Methods to:**
    *   View the access code (authorized users only)
    *   Update the access code after validating its length
    *   Display staff information
*   Use the `@property` decorator to manage the access code instead of traditional getter and setter methods

```python
class Staff:
    def __init__(self, s_name: str, access_code: str):
        """
        Initializes a Staff member.
        :param s_name: Public attribute for staff name
        :param access_code: Initial access code (will be validated and stored privately)
        """
        self.s_name = s_name
        # Using the setter property directly during initialization to enforce validation
        self.access_code = access_code

    @property
    def access_code(self) -> str:
        """
        Getter method using @property decorator.
        Allows authorized viewing of the access code.
        """
        # Simple authorization simulation (can be expanded if needed)
        # For this assignment, it safely returns the private attribute when accessed
        return self.__access_code

    @access_code.setter
    def access_code(self, new_code: str):
        """
        Setter method using @access_code.setter.
        Validates the length of the access code before updating.
        """
        # Basic validation: ensuring code is not empty and meets a standard length requirement (e.g., 4 or more characters)
        if len(new_code) < 4:
            raise ValueError("Security Alert: Access code is too short! It must be at least 4 characters long.")

        self.__access_code = new_code
        print(f"[Success] Access code updated successfully for {self.s_name}.")

    def display_staff_info(self):
        """
        Displays the staff information.
        """
        print("\n--- Staff Profile ---")
        print(f"Staff Name : {self.s_name}")
        print(f"Access Code: {self.access_code}")
        print("---------------------")


# --- Demonstration of the Class Functionality ---
if __name__ == "__main__":
    print("--- Creating a new Staff member ---")
    staff1 = Staff("John Doe", "UMaT2026")
    staff1.display_staff_info()

    print("\n--- Updating Access Code (Valid) ---")
    staff1.access_code = "NewSecureCode99"
    staff1.display_staff_info()

    print("\n--- Attempting Invalid Access Code Update (Too Short) ---")
    try:
        staff1.access_code = "123"
    except ValueError as e:
        print(e)
```
