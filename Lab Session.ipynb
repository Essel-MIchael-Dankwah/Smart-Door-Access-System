# Case Study Activity: Smart Door Access System for UMaT

You have been contacted by the **Dean of ORIC, UMaT**, to develop a smart door access system for UMaT admin staff. Each staff has an **access code** that must be protected from direct modification.

### Requirements:
Create a `Staff` class with:
*   **Public attribute:** `s_name`
*   **Private attribute:** `__access_code`
*   **Methods to:**
    *   View the access code (authorized users only)
    *   Update the access code after validating its length
    *   Display staff information
*   Use the `@property` decorator to manage the access code instead of traditional getter and setter methods

```python
# Creating Class
class Staff:
    # Definig Attributes
    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code
    # Creating the Getters and Setters
    def view_staff_info(self):
        print(f"Staff Name : {self.s_name}")
    def code(self):
        return self.__access_code
    def update(self,value):
        if len(str(value)) == 4:
            self.__access_code = value
        else:
            print("Invalid Access Code")

    # Using the @property method
    @property
    def access_code(self):
        authorization = input("Enter password to view {self.s_name}'s access code   -->  ")
        if authorization == "1234":
            return self.__access_code
        else:
            return "Access Denied"
    
    @access_code.setter
    def access_code(self, value):
        if len(str(value)) == 4:
            self.__access_code = value
        else:
            print("Invalid access code, it should be exactly four charaters")



# Testing Functionality
staff1 = Staff("Essel", "2026") # Creating Staff Object

staff1.view_staff_info() # Viewing Staff information

staff1.access_code = "9876" # Updating the access code

staff1.access_code = "12" # Updating the access code with wrong requirements

print(staff1.access_code) # Printing the updated access code
```
